## Differential Privacy

```python
epsilons = [0.01,0.2,np.log(2),np.log(3)]

def add_laplace_noise(s, sensitivity, epsilon):
  return s + np.random.laplace(loc=0, scale=sensitivity/epsilon)

def percentage_error(orig, est):
  return ((orig-est) / orig) * 100
```

### Sensitivity for each analyse
#### Count

```python
def sensitivity_count(df,query):
    count_infringed = df.query(query).shape[0]
    values = [i for i in range(count_infringed,0,-1)]
    maxi = 0
    for ind, value in enumerate(values):
        if ind == 0:
            maxi = abs(count_infringed - value)
        else:
            aux = abs(count_infringed - value)
            if aux > maxi:
                maxi = aux
    return count_infringed, maxi

def sensitivity_sum(df,query,column):

    df_query = df_raw_clean.query(query)[column]

    subsets_sums = [df_query[:i].sum() for i in range(df_query.shape[0],0,-1)]
    query_sum = df_query.sum()

    for ind, value in enumerate(subsets_sums):
        if ind == 0:
            maxi = abs(query_sum - value)
        else:
            aux = abs(query_sum - value)
            if aux > maxi:
                maxi = aux

    return query_sum,maxi

def sensitivity_mean(df,query,column):

    df_query = df.query(query)[column]
    
    query_mean = df_query.mean()
    subsets_mean = [df_query[:i].mean() for i in range(df_query.shape[0],0,-1)]

    for ind, value in enumerate(subsets_mean):
        if ind == 0:
            maxi = abs(query_mean - value)
        else:
            aux = abs(query_mean - value)
            if aux > maxi:
                maxi = aux
    
    return query_mean,maxi
```

```python
count_value,sens = sensitivity_count(df_raw_clean,'infringed == 1')

print('Sensitivity: ',sens)
print(f'Normal: {count_value}')

for eps in epsilons:
    dp_count = add_laplace_noise(count_value,sens,eps)
    print(f'EPSILON = {eps}\nDP: {dp_count}\nError: {percentage_error(count_value,dp_count)}\n')
```

```python
query = 'infringed == 1'
```

#### Sum

```python
col_sum = 'annual_income'

query_sum,sens = sensitivity_sum(df_raw_clean,query,col_sum)

print('Sensitivity: ',sens)
print(f'Normal: {query_sum}')
for eps in epsilons:
    dp_sum = add_laplace_noise(query_sum,sens,eps)
    print(f'EPSILON = {eps}\nDP: {dp_sum}\nError: {percentage_error(query_sum,dp_sum)}\n')
```

#### Mean

```python
col_mean = 'credit_amount'

mean_value,sens = sensitivity_mean(df_raw_clean,query,col_mean)

print('Sensitivity: ',sens)
print('Normal: ',mean_value)

for eps in epsilons:
    dp_mean = add_laplace_noise(mean_value,sens,eps)
    print(f'EPSILON = {eps}\nDP: {dp_mean}\nError: {percentage_error(mean_value,dp_mean)}\n')
```
