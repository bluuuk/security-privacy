## Differential Privacy

```python
epsilons = [0.01,0.2,np.log(2),np.log(3)]
epsilon_dict = {0.01:'0.01',0.2:'0.2',np.log(2):'log(2)',np.log(3):'log(3)'}


def add_laplace_noise(s, sensitivity, epsilon):
  return s + np.random.laplace(loc=0, scale=sensitivity/epsilon)

def percentage_error(orig, est):
  return ((orig-est) / orig) * 100
```

#### Count

```python
def dp_count_data(df,queries):

    df_columns = ['query','epsilon','sens','real_count','dp_count','error']
    df_count = pd.DataFrame(columns = df_columns)
    ind = 0
    for query in queries:
        df_query = df.query(query)
        original_count = df_query.shape[0]

        for i in range(len(epsilons)):
            dp_count = add_laplace_noise(original_count,1,epsilons[i])  
            error_margin = percentage_error(original_count,dp_count)
            df_count.loc[ind] = [query,epsilon_dict.get(epsilons[i]),1,original_count,dp_count,error_margin]
            ind+= 1
    
    df_count.to_csv('dp_count.csv',index=False)
    return df_count
```

```python
queries = ['infringed == 0','annual_income > 10000','credit_amount > 40000','days_employed > 10','goods_valuation > 200000']

dp_count_data(df_raw_clean,queries)
```

#### Sum

```python
def dp_sum_data(df,cols):

    df_columns = ['attribute','epsilon','sens','real_sum','dp_sum','error']

    ind = 0
    df_sum = pd.DataFrame(columns = df_columns)

    for col in cols:
        query_sum = df_raw_clean[col].sum()
        sens = round(df_raw_clean[col].max())

        for i in range(len(epsilons)):

            dp_sum = add_laplace_noise(query_sum,sens,epsilons[i])
            error_margin = percentage_error(query_sum,dp_sum)
            df_sum.loc[ind] = [col,epsilon_dict.get(epsilons[i]),sens,query_sum,dp_sum,error_margin]
            ind+= 1

    df_sum.to_csv('dp_sum.csv',index=False)
    return df_sum
```

```python
dp_sum_data(df_raw_clean,cols_sum)
```

#### Mean

```python
def dp_mean_data(df,cols):
    df_columns = ['attribute','epsilon','sens','real','dp_mean','error']
    ind = 0
    df_mean = pd.DataFrame(columns = df_columns)

    for col in cols:

        df_col = df_raw_clean[col]
        mean_value = df_col.mean()
        sum_value = df_col.sum()
        sens = round(df_col.max())

        for i in range(len(epsilons)):

            dp_mean = add_laplace_noise(sum_value,sens,epsilons[i]) / add_laplace_noise(df_raw_clean.shape[0],1,epsilons[i])
            error_margin = percentage_error(mean_value,dp_mean)
            df_mean.loc[ind] = [col,epsilon_dict.get(epsilons[i]),sens,mean_value,dp_mean,error_margin]
            ind+= 1
    df_mean.to_csv('dp_mean.csv',index = False)
    return df_mean
```

```python
dp_mean_data(df_raw_clean,cols_mean)
```
